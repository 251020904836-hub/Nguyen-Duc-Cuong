```python
n = int(input("Nhập số lượng phần tử: "))

tong_chan = 0
tong_le = 0

for i in range(n):
    x = int(input(f"Nhập phần tử thứ {i+1}: "))
    if x % 2 == 0:
        tong_chan += x
    else:
        tong_le += x

print("Tổng phần tử chẵn:", tong_chan)
print("Tổng phần tử lẻ:", tong_le)
```
